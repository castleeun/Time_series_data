```python
# %matplotlib widget
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('2023_no_age.csv')
```

```python
df
```

```python
df['날짜'] = pd.to_datetime(df['날짜'], format='%Y-%m-%d')
df['Days_Past'] = (df['날짜'] - pd.Timestamp('2023-01-01')).dt.days
```

```python
time_cols = df.columns.tolist()[4:28].copy()
print(time_cols)
```

```python
df['TOTAL_CNT'] = df[time_cols].sum(axis=1)
```

```python
df[df['역번호']==1104][df['구분']=='승차'].plot(x='Days_Past',y='TOTAL_CNT')
```

```python
df['day'] = df['날짜'].dt.weekday
```

```python
df
```

```python
result = (
    df
    .drop(['날짜','역명','Days_Past','TOTAL_CNT'],axis=1)
    .groupby(['역번호','구분','day']).mean()
    .reset_index()
)
# result[result['역번호']==1104]
```

```python
result
```

```python
def filter_by_day(feat,select='None'):
    if select=='weekday':
        return np.concatenate([feat[0:5],feat[7:12]])
    if select=='weekend':
        return np.concatenate([feat[5:7],feat[12:14]])
    else:
        return feat
def get_feat(select='None'):
    stations = range(1101,1123)
    feats = []
    implots = []
    for st in stations:
        row = result[result['역번호']==st][time_cols].to_numpy()
        row_norm = row/np.max(row,axis=1,keepdims=True)
        st_norm = row/np.max(row,keepdims=True)
        # feats.append(st_norm.flatten())
        feats.append(filter_by_day(st_norm,select).flatten())
        # implots.append(st_norm)
    feats = np.array(feats)
    # feats2 = feats/np.max(feats,axis=1,keepdims=True)
    print(feats.shape)
    return feats
```

```python
station_dict = {
    101:	"판암(대전대)",
    102:	"신흥",
    103:	"대동(우송대)",
    104:	"대전",
    105:	"중앙로",
    106:	"중구청",
    107:	"서대전네거리",
    108:	"오룡",
    109:	"용문",
    110:	"탄방",
    111:	"시청",
    112:	"정부청사(신협중앙회)",
    113:	"갈마",
    114:	"월평(한국과학기술원)",
    115:	"갑천",
    116:	"유성온천(충남대·목원대)",
    117:	"구암",
    118:	"현충원(한밭대)",
    119:	"월드컵경기장(노은도매시장)",
    120:	"노은",
    121:	"지족(침신대)",
    122:	"반석(칠성대)"
}
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X = pca.fit_transform(get_feat('weekend'))
idx = np.arange(X.shape[0])
colors = [
    'red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 
    'olive', 'cyan', 'gold', 'lime', 'magenta', 'teal', 'navy', 'maroon', 
    'turquoise', 'yellow', 'indigo', 'violet', 'peru', 'slategray'
]
plt.figure(figsize=(10, 10))
plt.title('주말 요일별 승하차 평균 이용객 수 패턴 PCA')
for i, data in enumerate(X):
    plt.scatter(X[i,0], X[i,1], label=station_dict[i+101], c=colors[i])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.xlim([-2,3])
plt.ylim([-2,3])
plt.legend(loc='best', bbox_to_anchor=(1, 1))
plt.show()
```
